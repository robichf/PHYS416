```python
#this program computes the function F(x) = x^a (x_f - x)^b/[x_m^a (x_f - x_m)^b] where 0 < x < x_f and a,b > 0
#  and F'(x_m) = 0

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

xf = 10
iscal = 1
a = 7*iscal
b = 12*iscal
npts = 1000
dx = xf/npts
xm = xf*a/(a+b)
fxm = pow(xm,a) * pow(xf-xm,b)
#Prob 2.3 g
#set the number of coins
func = np.empty(npts+1)
gaus = np.empty(npts+1)
xval = np.empty(npts+1)
deltax2 = 2.0*xf*xf*a*b/pow((a+b),3)
n = 0
while n < npts+1:
    x = n*dx
    xval[n] = x
    func[n] = pow(x,a) * pow(xf-x,b)/fxm
    gaus[n] = np.exp(-(x-xm)*(x-xm)/deltax2)
    n += 1
    
```

```python
plt.plot(xval,func,label='F(x)')
plt.plot(xval,gaus,label='exp(-(x-xm)^2/deltax^2)')
plt.legend()
plt.title('F(x) and Gaussian')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```
