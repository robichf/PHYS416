```python
#this program computes the partition function for a rotor as well as plotting the probability for the rotor to have
#  angular momentum l

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

#the energies of the rotor are E_l = epsilon l (l+1)
#betepsinv = 1/(beta X epsilon) = (k T)/epsilon is a scaled temperature
betepsinv=100.0
#estimate the maximum number of l needed
lmaxflt = 6.0*np.sqrt(betepsinv)
lmax = int(lmaxflt)
print(lmax)
prob = np.empty(lmax)
lval = np.empty(lmax)

#fill the values without the 1/Z term and compute the partition function
partfun = 0.0
l = 0
while l < lmax:
    lval[l] = 1.0*l
    prob[l] = (2*l+1)*np.exp(-l*(l+1)/betepsinv)
    partfun += prob[l]
    l+=1

#compute an approximation to the partion function
approxz = betepsinv+1.0/3.0
print('Z =',partfun,approxz)
#include the partition function to convert to probability
l=0
while l < lmax:
    lval[l] = 1.0*l
    prob[l] = (2*l+1)*np.exp(-l*(l+1)/betepsinv)/partfun
    l+=1

```

```python
plt.plot(lval,prob,label='prob')
plt.legend()
plt.title('Rotor: prob vs l')
plt.xlabel('l')
plt.ylabel('prob')
plt.show()
```
