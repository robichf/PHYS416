```python
#this program computes the probability that two systems of harmonic oscillators share a fixed number of excitations
#System A has N_A oscillators
#System B has N_B oscillators
#they share a total number of qtot excitations
#the goal is to find the probability that q excitations are in System A and qtot-q excitations are in System B

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

from math import comb

#set the values of N_A, N_B, and total number of excitations
iscale = 1
na = 5*iscale
nb = 10*iscale
qtot = 30*iscale

#initialize arrays with empty values to be filled in the loops below
omega_a = np.empty(qtot+1)
omega_b = np.empty(qtot+1)
omega_tot = np.empty(qtot+1)
prob_tot = np.empty(qtot+1)
qval = np.empty(qtot+1)

#first calculate the Omega_A and Omega_B for different numbers of excitation
iq = 0
while iq < qtot+1:
  omega_a[iq] = comb(iq+na-1,iq)
  omega_b[iq] = comb(iq+nb-1,iq)
  qval[iq] = iq*1.0
  #increment q
  iq += 1

#calculate the total microstates divied 
iq = 0
while iq < qtot+1:
  omega_tot[iq] = omega_a[iq]*omega_b[qtot-iq]
  prob_tot[iq] = omega_tot[iq]/comb(na+nb+qtot-1,qtot)
  if iq > 7*iscale and iq < 13*iscale:
      print(iq,'{:.3e}'.format(omega_a[iq]),'{:.3e}'.format(omega_b[qtot-iq]),'{:.3e}'.format(omega_tot[iq]),'{:.3e}'.format(prob_tot[iq]))
  #increment q
  iq += 1

```

```python
plt.plot(qval/qtot,prob_tot,label='probability for division of excitations')
plt.legend()
plt.xlabel('q_A')
plt.ylabel('prob')
plt.show()
```
