```python
#this program computes the number of microstates of a system with N oscillators and q excitations
#the goal is to show how the microstates change with increasing q for fixed N or for fixed q as a function of N

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

from math import comb

#set the number of oscillators
nosc = 20
#set the total number of q
qtot = 50

#initialize arrays with empty values to be filled in the loops below
omega = np.empty(qtot+1)
logomega = np.empty(qtot+1)
qval = np.empty(qtot+1)

#Calculate the number of microstates for different q
q = 0
while q < qtot+1:
  omega[q] = comb(q+nosc-1,q)
  logomega[q] = np.log(comb(q+nosc-1,q))
  qval[q] = q*1.0
  print(q,'{:.3e}'.format(omega[q]),'{:.3e}'.format(logomega[q]))
  #increment q
  q += 1
```

```python
plt.plot(qval/qtot,logomega,label='ln(Omega)')
plt.legend()
plt.xlabel('q/qtot')
plt.ylabel('ln(Omega)')
plt.show()
```
