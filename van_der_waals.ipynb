```python
#this program computes the van der waals variables for different temperatures
#  all of the parameters are scaled by their critical value except G which is divided by (2/3)(a/b)N

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

#set the scaled temperature
ttil = 0.8

#set the number of points in the plot
npts=1000

#set arrays for the functions
vtil = np.empty(npts+1)
gsca = np.empty(npts+1)
ptil = np.empty(npts+1)

dvtil = (10.0 - (1.0/3.0))/(npts+2)

# fill the arrays to be plotted
n = 0
while n < npts+1:
    vt = (1.0/3.0) + dvtil*(n+1)
    vtil[n] = vt
    ptil[n] = 8.0*ttil/(3.0*vt-1.0) - 3.0/(vt*vt)
    gsca[n] = (4.0*ttil/9.0/(3.0*vt-1.0))*(1.0-(3.0*vt-1.0)*np.log(3.0*vt-1.0)) - 1.0/vt
    n += 1
    
```

```python
plt.plot(vtil,ptil,label='P/P_c')
plt.legend()
ax = plt.gca()
ax.set_xlim([0,5.0])
ax.set_ylim([-0.5, 2.0])
plt.title('P vs V')
plt.xlabel('V/V_c')
plt.ylabel('P/P_c')
plt.show()
```

```python
plt.plot(vtil,gsca,label='G/[(2/3)N(a/b)]')
plt.legend()
ax = plt.gca()
ax.set_xlim([0,5.0])
ax.set_ylim([-2.0, 0.0])
plt.title('G vs V')
plt.xlabel('V/V_c')
plt.ylabel('G/[(2/3)N(a/b)]')
plt.show()
```

```python
plt.plot(ptil,gsca,label='G/[(2/3)N(a/b)]')
plt.legend()
ax = plt.gca()
ax.set_ylim([-1.5, -0.8])
ax.set_xlim([-0.5, 2.0])
plt.title('G vs P')
plt.xlabel('P/P_c')
plt.ylabel('G/[(2/3)N(a/b)]')
plt.show()
```
