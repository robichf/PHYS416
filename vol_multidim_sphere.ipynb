```python
#this program computes numerically finds the number of integers n_1^2 + n_2^2 + ... n_N^2 <= n^2
# this will do the calculations for N = 2, 3, 4 up to nmax
# this will do the calculation by numerically summing and by doing the approximate value from integration

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

pi = 2.0*np.asin(1.0)
nmax = 20

#set arrays for the function, the Gaussian approximation to the function, and the x-values
nval = np.empty(nmax+1)
vol2app = np.empty(nmax+1)
vol3app = np.empty(nmax+1)
vol4app = np.empty(nmax+1)
vol2exa = np.empty(nmax+1)
vol3exa = np.empty(nmax+1)
vol4exa = np.empty(nmax+1)

# fill the arrays to be plotted
n = 0
while n < nmax+1:
    nval[n] = n
    vol2app[n] = pi*pow(n,2)
    vol3app[n] = (4.0*pi/3.0)*pow(n,3)
    vol4app[n] = (pi*pi/2.0)*pow(n,4)
    sum2 = 0.0
    sum3 = 0.0
    sum4 = 0.0
    n1 = -n
    while n1 < n+1:
        n2=-n
        while n2 < n+1:
            if(n1*n1+n2*n2 <= n*n):
                sum2 += 1.0
            n3 = -n
            while n3 < n+1:
                if(n1*n1+n2*n2+n3*n3 <= n*n):
                    sum3 += 1.0
                n4 = -n
                while n4 < n+1:
                    if(n1*n1+n2*n2+n3*n3+n4*n4 <= n*n):
                        sum4 += 1.0
                    n4+=1
                n3+=1
            n2+=1
        n1+=1
    vol2exa[n] = sum2
    vol3exa[n] = sum3
    vol4exa[n] = sum4
    print(n,'{:.3e}'.format(vol2app[n]),'{:.3e}'.format(sum2),'{:.3e}'.format(vol3app[n]),'{:.3e}'.format(sum3),'{:.3e}'.format(vol4app[n]),'{:.3e}'.format(sum4))
    n += 1
    
```

```python
plt.plot(nval,vol2exa,label='V_2 exact')
plt.plot(nval,vol2app,label='V_2 approx')
plt.legend()
plt.title('D=2 volume')
plt.xlabel('n')
plt.ylabel('V_3')
plt.show()
```

```python
plt.plot(nval,vol3exa,label='V_3 exact')
plt.plot(nval,vol3app,label='V_3 approx')
plt.legend()
plt.title('D=3 volume')
plt.xlabel('n')
plt.ylabel('V_2')
plt.show()
```
