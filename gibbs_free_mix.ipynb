```python
#this program computes the Gibbs free energy for a mixture
# it uses an idealized mixing function which is quadratic in the fraction x

#numpy allows for most basic math, including complex numbers
import numpy as np
#pyplot is the 'default' choice for plotting in python
import matplotlib.pyplot as plt

#Define some parameters
g0a = -100.0e3
g0b = -99.0e3
rt = 8.31*298
deluhalf = 2.0e3
g = g0a
x=0.0
print('{:.2e}'.format(x),'{:.4e}'.format(g))
x=0.2
while x < 0.9:
    g = (1.0-x)*g0a + x*g0b + rt*(x*np.log(x)+(1.0-x)*np.log(1.0-x)) + deluhalf*4.0*x*(1.0-x)
    print('{:.2e}'.format(x),'{:.4e}'.format(g))
    x += 0.2
x = 1.0
g = g0b
print('{:.2e}'.format(x),'{:.4e}'.format(g))

#fill an array to plot g(x)
npts=1000
gfun = np.empty(npts+1)
xfun = np.empty(npts+1)
dx = 1.0/npts
x = 0.0
n = 0
x = dx*n
gfun[n] = g0a
xfun[n] = x
while n < npts:
    x=dx*n
    xfun[n] = x
    gfun[n] = (1.0-x)*g0a + x*g0b + rt*(x*np.log(x)+(1.0-x)*np.log(1.0-x)) + deluhalf*4.0*x*(1.0-x)
    n+=1

n=npts
x = 1.0
gfun[n] = g0b
xfun[n] = x
```

```python
plt.plot(xfun,gfun/1.e3,label='G(x)')
plt.legend()
plt.title('Gibbs free energy vs fraction')
plt.xlabel('x')
plt.ylabel('G (kJ/mol)')
plt.show()
```
